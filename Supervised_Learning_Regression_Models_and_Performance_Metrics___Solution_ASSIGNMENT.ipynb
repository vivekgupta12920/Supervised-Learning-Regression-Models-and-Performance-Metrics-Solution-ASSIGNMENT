Q6: Name and briefly describe three common evaluation metrics for regression
models?


-> Mean Absolute Error (MAE): This metric calculates the average magnitude of errors between actual and predicted values without considering their direction. It treats all errors equally, making it more robust to outliers than other metrics. It is highly interpretable because it is expressed in the same units as the target variable.

Root Mean Squared Error (RMSE): RMSE is the square root of the average of squared differences between predicted and actual values. Like MAE, it is expressed in the original units of the data, but because it squares the errors before averaging, it penalizes larger errors more heavily. It is widely used in 2025 as a standard measure of a model's prediction precision.

R-squared (\(R^{2}\)): Also known as the coefficient of determination, this metric measures the proportion of variance in the dependent variable that is explained by the independent variables. It ranges from 0 to 1, where 1 indicates a perfect fit. It is primarily used to evaluate the overall goodness-of-fit and explanatory power of a model rather than direct prediction error

Q8: Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept?


```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Prepare the data
# X must be a 2D array (samples, features)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

# 2. Initialize and fit the model
model = LinearRegression()
model.fit(X, y)

# 3. Access and print the slope and intercept
# .coef_ returns an array of slopes for each feature
# .intercept_ returns the y-intercept as a float
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (Coefficient): {slope:.4f}")
print(f"Intercept: {intercept:.4f}")

# Optional: Predicting a new value
new_x = np.array([[6]])
prediction = model.predict(new_x)
print(f"Prediction for X=6: {prediction[0]:.4f}")
```
